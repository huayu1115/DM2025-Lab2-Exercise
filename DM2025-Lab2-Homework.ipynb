## 1. Preprocessing Steps

```python
# ËÆÄcsv
import os
import pandas as pd
import json

data_identification = pd.read_csv("data\competition\data_identification.csv")
emotion = pd.read_csv("data\competition\emotion.csv")
samplesubmission = pd.read_csv("data\competition\samplesubmission.csv")
```

```python
# ËÆÄ json
with open(r"data\competition\final_posts.json", 'r', encoding='utf-8') as f:
    data = json.load(f)

records = []
for item in data:
    root = item.get("root", {})
    _type = root.get("_type")
    post_info = root.get("_source", {}).get("post", {})

    record = {
        "_type": _type,
        "post_id": post_info.get("post_id"),
        "text": post_info.get("text"),
        "hashtags": post_info.get("hashtags"),
    }
    records.append(record)

df = pd.DataFrame(records)
df.head()
```

```python
train_df = pd.merge(
    emotion,
    df,
    left_on='id',      
    right_on='post_id',
    how='inner'       
).drop(columns=['post_id'])
print(len(train_df))
train_df.head()
```

```python
test_df = pd.merge(
    samplesubmission,
    df,
    left_on='id',      
    right_on='post_id',
    how='inner'       
).drop(columns=['post_id'])
print(len(test_df))
test_df.head()
```

## 3. Model Implementation Steps

```python
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
from torch.nn import CrossEntropyLoss
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, classification_report
import numpy as np
```

```python
print("PyTorch version:", torch.__version__)
print("CUDA available:", torch.cuda.is_available())
print("GPU count:", torch.cuda.device_count())

if torch.cuda.is_available():
    print("GPU name:", torch.cuda.get_device_name(0))
```

ÊÉÖÁ∑íÊ®ôÁ±§Á∑®Á¢º

```python
label2id = {"joy":0, "anger":1, "surprise":2, "sadness":3, "fear":4, "disgust":5}
id2label = {v:k for k,v in label2id.items()}
train_df["label_id"] = train_df["emotion"].map(label2id)
```

ÂàÜÂâ≤Ë®ìÁ∑¥ÈõÜËàáÈ©óË≠âÈõÜÔºåÁî®ÂàÜÂ±§ÊäΩÊ®£ stratified samplingÔºå‰øùÊåÅÂéüÊú¨Ê®ôÁ±§ÁöÑÊØî‰æãÂàÜÂ∏É

```python
train_texts, val_texts, train_labels, val_labels = train_test_split(
    train_df["text"].tolist(),
    train_df["label_id"].tolist(),
    test_size=0.2,
    random_state=42,
    stratify=train_df["label_id"]
)

test_texts = test_df["text"].tolist()
```

Âª∫Á´ã roberta-base Tokenizer

```python
model_name = "roberta-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)

MAX_LEN = 200
def tokenize(texts):
    return tokenizer(texts, padding="max_length", truncation=True, max_length=MAX_LEN, return_tensors="pt")
```

Âª∫Á´ã Dataset: tokenize Âè•Â≠ê

```python
class EmotionDataset(Dataset):
    def __init__(self, texts, tokenizer, labels=None):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = self.tokenizer(
            self.texts[idx],
            padding="max_length",
            truncation=True,
            max_length=MAX_LEN,
            return_tensors="pt"
        )
        
        item = {key: val.squeeze(0) for key, val in encoding.items()}

        # only include labels if available
        if self.labels is not None:
            item["labels"] = torch.tensor(self.labels[idx])

        return item
```

```python
train_dataset = EmotionDataset(train_texts, tokenizer, labels=train_labels)
val_dataset = EmotionDataset(val_texts, tokenizer, labels=val_labels)
test_dataset = EmotionDataset(test_texts, tokenizer)  # labels=None
```

```python
train_dataset[0].keys()
```

Âà©Áî® class weights ËôïÁêÜ‰∏çÂπ≥Ë°°

```python
counts = train_df["label_id"].value_counts().sort_index().tolist()  # Êåâ label_id ÊéíÂ∫è
weights = [sum(counts)/c for c in counts]
weights = torch.tensor(weights).float()
print(weights)
```

‰ΩøÁî® GPU Âä†ÈÄü

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

‰ΩøÁî®È†êË®ìÁ∑¥Ê®°Âûã

```python
model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=6
).to(device)
```

Ë®≠ÂÆöÂä†Ê¨äÊêçÂ§±ÂíåË®ìÁ∑¥ÂèÉÊï∏

```python
loss_fn = CrossEntropyLoss(weight=weights)

training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    logging_dir='./logs',
    learning_rate=2e-5,
    load_best_model_at_end=True,
    metric_for_best_model="f1"
)
```

Metrics

```python
def compute_metrics(p):
    preds = np.argmax(p.predictions, axis=1)
    f1 = f1_score(p.label_ids, preds, average="macro")
    return {"f1": f1}
```

Trainer

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

È©óË≠âÈõÜË©ï‰º∞


```python
val_output = trainer.predict(val_dataset)
preds = np.argmax(val_output.predictions, axis=1)
print("Classification Report:\n")
print(classification_report(val_labels, preds, target_names=list(label2id.keys())))
```

```python
test_output = trainer.predict(test_dataset)
predictions = np.argmax(test_output.predictions, axis=1)
pred_labels = [id2label[i] for i in predictions]
print(pred_labels)
```

```python
submission_df = pd.DataFrame({
    "id": test_df.id, 
    "emotion": pred_labels
})

# Save to CSV
submission_df.to_csv("test_predictions.csv", index=False)
```
